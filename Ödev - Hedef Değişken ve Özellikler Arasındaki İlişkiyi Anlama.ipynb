```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt

weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

```python
weather_df["TemperatureDifference"] = weather_df["Apparent Temperature (C)"] - weather_df["Temperature (C)"]
weather_df.head()

#Hedef değişken negatif değerlere sahip, bu modeli etkiler mi ?
X = weather_df[["Humidity", "Wind Speed (km/h)"]]
Y = weather_df["TemperatureDifference"]

lrm = linear_model.LinearRegression()
lrm.fit(X,Y)

print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
#Statsmodel ile yeni bir model oluşturma
import statsmodels.api as sm
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()

```

Model katsayılarının P değerleri 0 olduğundan istatistiksel olarak anlamlıdır.( İstatistiksel olarak anlamlı olması tam olarak neyi ifade eder, neyi açıklar? )

Özelliklerin değerindeki artış, katsayılar negatif değer olduğundan hedef değerde azalış anlamına gelir. (Görünür sıcaklık farkı azalır demek).

Windspeed'in katsayısı sıfıra yakın (0.1), bu yüzden varyansın az kısmını taşıdığını (bilginin az kısmını taşıdığını düşünüyorum.)

Humidity'nin katsayısı daha yüksek olduğundan bilginin coğunluğunun bu özellikte olduğunu, varyansın bu özellikte daha çok olduğunu düşünüyorum

Bir önceki modelde R squared 0.44'tü hatırladığım kadarıyla, bu modelimizde R squared 0.288. Yani model daha az hataya sahip ve daha iyi bir model ?

```python
#Ardından, yukarıdaki modele nem ve rüzgar hızı etkileşimini dahil edin ve OLS'yi kullanarak modeli tahmin edin. Katsayılar istatistiksel olarak anlamlı mıdır? 
#Nem ve rüzgar hızı için tahmini katsayıların işaretleri değişti mi? Tahmin edilen katsayıları yorumlayınız.

weather_df["Hum_and_Wind"] = weather_df["Humidity"] * weather_df["Wind Speed (km/h)"]

X = weather_df[["Humidity", "Wind Speed (km/h)", "Hum_and_Wind"]]
Y = weather_df["TemperatureDifference"]

X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

Yorumlarımı yaptım ama katsayıların yorumlanmasını yinede beraber yapabilir miyiz.

Sabitimizin P değeri artmış görünüyor, ancak yinede 0.1'den küçük olduğundan istatistiksel olarak hala anlamlı.

Nem ve rüzgar özelliklerinin bireysel katsayısı azalmış, yeni oluşturduğumuz etkileşimli özelliğin katsayısı ise bunlara göre oldukça yüksek görünüyor. Hedef değişkenindeki varyansın çoğunu bu yeni özellik açıklıyor. ( Yorumum doğru ise final modeli oluştururken nem ve rüzgar hızı özelliğini çıkarıp, modele sadece bunların etkileşimli yeni oluşturduğumuz özelliğini ekleyerek mi devam etmek gerekir ?)

Etkileşimli özelliğin P değeri 0 olduğundan istatistiksel olarak anlamlı.( Sadece p değerine bakarak eğer p değeri yüksekse o zaman o özelliğin katsayısı sıfıra yakın oldugundan, model üzerinde aslında pek bir etkiye sahip olmadığını düşünüp başka birşeye bakmadan, direkt modelden cıkarabilir miyiz ? )
