**(1)** Plot a bar chart for total cases of the 20 countries that havebiggest numbers.

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
df1 = df[df.date == "2021-03-03"].sort_values("total_cases")
df2= df1[-40:-20]
```

```python
plt.figure(figsize=(10, 5))

plt.bar(df2.location, df2.total_cases, color = "#FF914D")
plt.xlabel("Countries")
plt.ylabel("Total Case Numbers")
plt.xticks(rotation= 90)
plt.show()
```

**(3)** Plot a scatter plot of new cases and new death for Germany and France.

```python
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
plt.title("New Cases ve Total Deaths")

for locations in ['Germany', 'France']:
    plt.scatter(df[df.location == locations].new_cases, 
                df[df.location == locations].total_deaths, label = locations)

plt.xlabel('New_Cases')
plt.ylabel('Total_Deaths')
plt.legend()
    
plt.show()
```

**(4)** Plot a boxplot for daily deaths for any country you choose.
