```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
df[:3]
```

```python
df.loc[(df.location.isin(["Spain", "France", "Germany", "Italy"])) & (df.date > "2020-02-24")] # burada date kolonu string değil mi, ben ">"(büyüktür) operatörünün burada
#çalısmasını beklemiyordum.
```

```python
df1 = df[df.location.isin(["Spain", "France", "Germany", "Italy"])].date == "2020-02-24"
```

```python
df2 =df.loc[(df.location.isin(["Spain", "France", "Germany", "Italy"])) & (df.date > "2020-02-24")] # burada date kolonu string değil mi, ben ">"(büyüktür) operatörünün burada
#çalısmasını beklemiyordum.
```

```python
plt.figure( figsize = (17,6), dpi=500)
sns.lineplot(x="date", y="total_deaths", data = df2, hue="location")
plt.title("Daily Total Deaths")
plt.xlabel("Dates", rotation=45)
plt.ylabel("Daily Deaths")
plt.show()
```

```python
plt.figure(figsize =(10,4), dpi=250)
sns.barplot(x="continent", y="total_deaths",  data=df, palette="Set2")
plt.title("Continents Average Deaths")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(8,3))

plt.subplot(2,2,1)
sns.distplot(df[(df.location == "Germany") % (df.new_deaths.dropna())]) 
plt.show()

##çalısmasını bekliyordum.
```

```python
plt.figure(figsize=(8,3), dpi = 300)

plt.subplot(131)
sns.boxplot(x="location", y="new_deaths", data=df.loc[df.location.isin(["Germany", "France"])])
plt.title("Boxplot")

plt.subplot(132)
sns.swarmplot(x="location", y="new_deaths", data=df.loc[df.location.isin(["Germany", "France"])])
plt.title("Swarmlot")

plt.subplot(133)
sns.violinplot(x="location", y="new_deaths", data=df.loc[df.location.isin(["Germany", "France"])])
plt.title("Boxplot")
```
